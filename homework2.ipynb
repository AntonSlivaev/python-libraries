```python
t = np.linspace(0, 15, 100)
```

```python
t
```

```python
f = np.cos(t)
```

```python
f
```

```python
plt.plot(t,f, color = 'green')
plt.title('График f(t)',fontsize = 14, loc = 'left')
plt.xlabel('Значения t')
plt.ylabel('Значения f')
plt.axis([0.5, 9.5, -2.5, 2.5])
```
